```python
# to see skewness 
df['Review Rating'].skew() 
```

```python
# why median not mean though there are no outliers
'''
Median imputation was chosen as a robust and practical 
approach for bounded rating data, 
even in the absence of skewness or outliers.''',
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.age.unique()
```

```python
# creating a age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df.frequency_of_purchases.unique()
```

```python
#create a column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly' : 7,
    'Monthly':30,
    'Quarterly': 90,
    'Bi-weekly':14,
    'Annually': 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).astype('Int64')
```
